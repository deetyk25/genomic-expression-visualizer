```python
# load cleaned data
import pandas as pd
from src.visualize import plot_top_variable_genes

df_log = pd.read_csv("../data/expression_cleaned_log.csv", index_col=0)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# transpose so rows = samples, columns = genes
X = df_log.T

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print("Explained variance:", pca.explained_variance_ratio_)
```

```python
labels = X.index

colors = ["red" if "INSM" in s else "blue" for s in labels]

plt.figure()
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=colors)

for i, label in enumerate(labels):
    plt.text(X_pca[i, 0], X_pca[i, 1], label)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Gene Expression Samples")
plt.show()
```

PCA was used to reduce high-dimensional gene expression data into a low-dimensional
representation. The first two principal components explained approximately 63.6%
and 26.0% of the total variance, respectively, capturing nearly 90% of the overall
variation.

Visualization of samples in PCA space revealed structure associated with
experimental condition, indicating that global gene expression patterns differ
between samples. This suggests the presence of biologically meaningful signal
suitable for downstream modeling and analysis.
